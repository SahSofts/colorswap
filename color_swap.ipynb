```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def color_swap(image_path, color_from, color_to, tolerance, opacity):
    img = Image.open(image_path).convert("RGBA")
    img_array = np.array(img)
    distances = np.linalg.norm(img_array[:, :, :3] - color_from, axis=2)
    mask = distances <= tolerance
    modified_img_array = np.where(mask[..., np.newaxis], [*color_to, opacity], img_array)
    modified_img = Image.fromarray(modified_img_array.astype(np.uint8))

    return modified_img

def process_image(image_path, color_from, color_to, tolerance, opacity):
    img = Image.open(image_path)
    modified_img = color_swap(image_path, color_from, color_to, tolerance, opacity)
    fig, axes = plt.subplots(1, 2, figsize=(7, 5))
    axes[0].imshow(img)
    axes[0].set_title('Before')
    axes[0].axis('off')
    axes[1].imshow(modified_img)
    axes[1].set_title('After')
    axes[1].axis('off')
    plt.show()

image_path = 'tanjiro.png'
color_from = [152, 25, 63]
color_to   = [255, 255, 255]
tolerance  = 100
opacity    = 255

process_image(image_path, color_from, color_to, tolerance, opacity)


image_path = 'zoro.png'
color_from = [129, 225, 26]  
color_to   = [217, 113, 255]  

tolerance = 50  
opacity   = 255    

process_image(image_path, color_from, color_to, tolerance, opacity)



```
